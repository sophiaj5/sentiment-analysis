```python
#imports
import pandas as pd
import numpy as np
import matplotlib as plt

from nltk.stem import WordNetLemmatizer
```

```python
#pulling in cleaned data
df = pd.read_csv('../dataset/cleaned_data')
df
```

```python
#function to preprocess the text
def preprocess(text_column):
    preprocessed_text = []
    for post in text_column:
        post = post.lower() #turns text into lowercase
        preprocessed_text.append(post)
    
    return preprocessed_text
```
